# Question 1 et 2 

```python
def draw_my_function(f, start, end, number_of_points):
    x_values = numpy.linspace(start, end, number_of_points)
    y_values = f(x_values)
    
    df_dx = numpy.gradient(y_values, x_values)
    
    plt.figure(figsize=(15,6))
    plt.plot(x_values, y_values  ,label="Ma Function")
    plt.plot(x_values, df_dx, label="La dérivée")
    plt.legend()

    plt.show()
    return x_values, y_values, df_dx
```

```python
start, end = 1, 10
number_of_points = 10000
f = lambda x_values : numpy.exp(-x_values/10)*numpy.sin(x_values)
```

```python
def get_information_for_function(x_values, y_values, slice_start, slice_end):
    
    y_values_new = y_values[(x_values>=slice_start) & (x_values<=slice_end)]
    
    print("la moyenne est de :", y_values_new.mean())  
    print("et l'écart type :", y_values_new.std())
    
```

```python
x_values, y_values, df_dx = draw_my_function(f, start, end, number_of_points)
```

# 3) Moyenne et écart Type 

```python
get_information_for_function(x_values, y_values, slice_start=3, slice_end=7)
```
